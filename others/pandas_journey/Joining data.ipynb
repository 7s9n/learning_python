```python
import pandas as pd
import numpy as np
```

```python
employees = {
    'last_name':[ 'Hussein' , 'Moataz' , 'Ekram' , 'Ahmed' , 'Shekah' ],
    'dept_id' : [1 , 2 , 3 , 4 , np.nan],
}
```

```python
departments = {
    'dept_id' : [1 , 2 , 3 , 4 , 5] ,
    'dept_name': [ 'HR' , 'IT' , 'Administration'  , 'Marketing' , 'Sales'],
}
```

```python
df1 = pd.DataFrame(employees)
```

```python
df1.head()
```

```python
df2 = pd.DataFrame(departments)
```

```python
df2.head()
```

```python
pd.merge( df1  , df2 , on='dept_id' , how='inner' ) #inner join
# pd.merge( df1  , df2) #Ok , because there's a common column dept_id
```

```python
pd.merge( df1  , df2 , on='dept_id' , how='left' ) #left outer join
# pd.merge( df1  , df2 , how='left' ) #Ok
```

```python
pd.merge( df1  , df2 , on='dept_id' , how='right' ) #right outer join
# pd.merge( df1  , df2 , how='right' ) #Ok
```

```python
pd.merge( df1  , df2 , on='dept_id' , how='outer' ) #full outer join
# pd.merge( df1  , df2 , how='outer' ) #Ok
```

```python
employees = {
    'last_name':[ 'Hussein' , 'Moataz' , 'Ekram' , 'Ahmed' , 'Shekah' ],
    'dept_id' : [1 , 2 , 3 , 4 , np.nan],
}
```

```python
departments = {
    'department_id' : [1 , 2 , 3 , 4 , 5] ,
    'dept_name': [ 'HR' , 'IT' , 'Administration'  , 'Marketing' , 'Sales'],
}
```

```python
df1 = pd.DataFrame(employees)
```

```python
df2 = pd.DataFrame(departments)
df2.set_index(['department_id'] , inplace=True)
```

```python
pd.merge( df1 , df2  , left_on='dept_id' , right_index=True) #inner join
```

```python
pd.merge( df1 , df2  , left_on='dept_id' , right_index=True , how='left') #left outer join
```

```python
pd.merge( df1 , df2  , left_on='dept_id' , right_index=True , how='right') #right outer join
```

```python
pd.merge( df1 , df2  , left_on='dept_id' , right_index=True , how='outer') #full outer join
```
