```python
import pandas as pd
import numpy as np
```

```python
people = {
    'first': ['Hussein' , 'Moataz' , 'Ekram' , 'Ahmed' , np.nan , None , 'NA'],
    'last': ['Sarea' , 'Sarea' , 'Ahmed' , 'Mohammed' , np.nan , np.nan , 'Missing'],
    'email': ['zzzsssx0@gmail.com' , 'zzzxxxx@yahoo.com' , 'eee11@gmail.com' , None , np.nan , 'zzzoo@hotmail.com' , 'NA'],
    'age': [22 , 21 , 17 , 66 , None , None , 'Missing'],
}
```

```python
df = pd.DataFrame(people)
```

```python
df.replace(['NA' , 'Missing'] , np.nan , inplace=True)
```

```python
df
```

```python
# df.dropna(axis='index', how='any') #default argument
```

```python
df.dropna()
```

```python
df.dropna(how='all')
```

```python
# df.dropna(axis='columns' , how='all')
# df.dropna(axis='columns' , how='any') # give us an empty dataframe
```

```python
df
```

```python
df.dropna(axis='index' , how='any' , subset=['email']) #if an email column isn't filled then we drop the row
```

```python
df.dropna(axis='index' , how='all' , subset=['last' , 'email']) 
```

```python
df.dropna(axis='index' , how='any' , subset=['last' , 'email']) 
```

```python
df.isna()
```

```python
# df.fillna('Missing')
df.fillna(0 , inplace=False)
```

```python
df['age'].fillna(0 , inplace=True)
df['age'] = df['age'].astype(np.uint8 )
```
